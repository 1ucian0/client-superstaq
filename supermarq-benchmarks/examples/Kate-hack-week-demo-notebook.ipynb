```python
%load_ext autoreload
%autoreload 2

import numpy as np
import supermarq
from supermarq.benchmarks.vqe_proxy import VQEProxy
from supermarq.benchmarks.vqe_hwea_clifford import VQE_HWEA_Clifford
import time

num_qubits = 12
```

```python
start_time = time.time()
test_vqe = VQEProxy(num_qubits,1)
end_time = time.time()
print(f'total time: {end_time-start_time}')

for i in range(len(test_vqe.circuit())):
    print(i)
    print(test_vqe.circuit()[i])
    print()
```
